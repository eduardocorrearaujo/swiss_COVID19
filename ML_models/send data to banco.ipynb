```python
import pandas as pd 
from sqlalchemy import create_engine

engine = create_engine("postgresql://epigraph:epigraph@localhost:5432/epigraphhub")
```

```python
def load_df(filename):    
    df = pd.read_csv(filename)

    del df['Unnamed: 0']
    #df.set_index('date', inplace = True)
    df.date = pd.to_datetime(df.date)
    
    return df

def save_to_database(df, table_name):
    df.to_sql(table_name, engine, schema= 'switzerland', if_exists = 'replace')
```

```python
df_icu1 = load_df('ICU_all_cantons.csv')

df_icu2 = load_df('ICU_all_cantons2.csv')

df_icu = pd.concat([df_icu1, df_icu2])

df_icu
```

```python
save_to_database(df_icu, 'ml_for_icu_all_cantons')
```

```python
df_hosp1 = load_df('total_hosp_all_cantons1.csv')

df_hosp2 = load_df('total_hosp_all_cantons2.csv')

df_hosp = pd.concat([df_hosp1, df_hosp2])

df_hosp
```

```python
save_to_database(df_hosp, 'ml_for_total_all_cantons')
```
