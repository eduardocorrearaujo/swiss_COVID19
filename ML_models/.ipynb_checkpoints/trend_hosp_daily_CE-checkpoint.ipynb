# LightGBM classifier for COVID trend
The need to predict COVID incidence trends has led us to try to forecast trends based on available data, cases hospitalizations, and deaths.
In this notebook was considered the data aggregate by week. 

## Loading the data 

```python
# importando as bibliotecas necessárias 
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
import matplotlib.pyplot as plt
import optuna.integration.lightgbm as lgbt  # Tuner for lgbm
from flaml import AutoML
import time 
import numpy as np 

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
def dataframe(estado):
    
    #o estado deve ser passado na forma de uma string do UF 
    
    df = pd.read_csv('dados_SRAG.csv', sep =',')
    
    # filtrando os casos de covid apenas
    df = df.loc[df.CLASSI_FIN == 5]
    
    df = df.loc[df.SG_UF_NOT == estado]
    
    df_hosp = df.copy()
    
    df_hosp['date'] = df_hosp.DT_SIN_PRI
    
    df_hosp.set_index('date', inplace = True)
    
    df_hosp.index = pd.to_datetime(df_hosp.index)
    
    df_hosp.sort_index(inplace = True)
    
    # agrupando os dados por dia 
    
    df_h_day  = df_hosp.resample('D', label = 'left', closed='left').count()

    df_h_d =  pd.DataFrame()

    df_h_d['daily_hosp'] = df_h_day.SG_UF_NOT
    
    # agora vamos trabalhar com os dados de óbitos
    
    df_deaths = df.copy()
    
    # vamos filtrar os óbitos
    
    df_deaths = df_deaths.loc[df_deaths.EVOLUCAO == 2]
    
    df_deaths['date'] = df_deaths.DT_EVOLUCA
    
    df_deaths.set_index('date', inplace = True)
    
    df_deaths.index = pd.to_datetime(df_deaths.index)
    
    df_deaths.sort_index(inplace = True)
    
    # agrupando os dados por dia 
    
    df_d_day  = df_deaths.resample('D', label = 'left', closed='left').count()

    df_d_d =  pd.DataFrame()

    df_d_d['daily_deaths'] = df_d_day.SG_UF_NOT
    
    
    # vamos agora criar o dataframe com os dados necessários para o modelo de tendências: 
    
    df_fil = pd.DataFrame()
                    
    df_fil['obitos_sivep'] = df_d_d.daily_deaths.astype('float')
    
    df_fil['srag_covid'] = df_h_d.daily_hosp.astype('float')
    
    

    
    return df_fil


df = dataframe('CE')

df.head(5)
```

Plotando os gráficos

```python
# calculando a média movel do dataframe:
m_movel = df.rolling(window=7).mean()

m_movel.dropna(inplace = True);

m_movel.head(8)
```

## The model
We will fit a LGBM classifier to model the odds of the incidence trend at a future date $T>t$ to be positive, where $t$ is the present date. As predictor we will user previous observations  up to $d$ time steps. the trend will simply be defined as $$trend = sign(H_T-H_t)$$

```python
def build_lagged_features(dt, maxlag=2, dropna=True):
    '''
    Builds a new DataFrame to facilitate regressing over all possible lagged features
    :param dt: Dataframe containing features
    :param maxlag: maximum lags to compute
    :param dropna: if true the initial rows containing NANs due to lagging will be dropped
    :return: Dataframe
    '''
    if type(dt) is pd.DataFrame:
        new_dict = {}
        for col_name in dt:
            new_dict[col_name] = dt[col_name]
            # create lagged Series
            for l in range(1, maxlag + 1):
                new_dict['%s_lag%d' % (col_name, l)] = dt[col_name].shift(l)
        res = pd.DataFrame(new_dict, index=dt.index)

    elif type(dt) is pd.Series:
        the_range = range(maxlag + 1)
        res = pd.concat([dt.shift(i) for i in the_range], axis=1)
        res.columns = ['lag_%d' % i for i in the_range]
    else:
        print('Only works for DataFrame or Series')
        return None
    if dropna:
        return res.dropna()
    else:
        return res
```

As mentioned before, we'll use the available data along with past data up to $d$ days specified in the `max_lag` variable below. We will also add a Trend. The regression model will be trained on the historical series up to 30 days ago so that we can be sure that most cases have already been reported.

```python
max_lag = 7
T = 7 
end = 30
data = m_movel[['srag_covid', 'obitos_sivep',]]
target_col = f'trend_{T}'

df_lag = build_lagged_features(data, maxlag=max_lag)
trend = [0 if t<=0 else 1 for t in np.sign(df_lag.srag_covid.shift(-T)-df_lag.srag_covid)]
df_lag[f'trend_{T}'] = trend#makes down trend 0 
df_lag = df_lag.dropna()
target = df_lag[target_col].copy()

df_lag
```

```python
fig, [a1,a2] = plt.subplots(2,1,figsize = (10,6))
df_lag.srag_covid.plot(ax=a1, label='incidence')
a1.grid()
df_lag.trend_7.plot.bar(ax=a2, label='trend');
# a2.grid()
plt.legend();
```

```python
del df_lag[target_col]
```

Dividindo o dataset em dados de treino e dados de teste


```python
X = df_lag
y = target

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, shuffle = False)
```

```python
inicio = time.time()

def tune_model(data, target, estimator='lgbm', task='quantile'):
    automl = AutoML()
    if estimator == 'lgbm':
        automl.fit(data, target, task="classification", estimator_list=["lgbm"])
    else:
        automl.fit(data, target, task="classification")
    return automl
        
automl = tune_model(X_train, y_train, estimator='lgbm')
automl.model

fim = time.time()
print(fim - inicio)
```

```python
''' retrieve best config and best learner'''
print('Best ML leaner:', automl.best_estimator)
print('Best hyperparmeter config:', automl.best_config)
print('Best accuracy on validation data: {0:.4g}'.format(1-automl.best_loss))
print('Training duration of best run: {0:.4g} s'.format(automl.best_config_train_time))
```

```python
''' pickle and save the automl object '''
import pickle
with open('trend_model.pkl', 'wb') as f:
    pickle.dump(automl, f, pickle.HIGHEST_PROTOCOL)
```

```python
''' compute predictions of training and testing dataset ''' 
y_pred = automl.predict(X_train)
y_pred_test = automl.predict(X_test)
#print('Predicted labels', y_pred)
#print('True labels', y_train)
```

### Sensitivity and Specificity
Sensitivity is defined as the recall of the posititive class, $TP/P$. Specificity on the other hand, can be thought of as the recall of the negative class, $$\frac{TN}{N}=1-FPR$$ or the complement of the false positive rate. So they can be easily obtained from the classification report table below

```python
print(classification_report(y_train, y_pred))
```

So, for the training set both are 1.

```python
print(classification_report(y_test, y_pred_test))
```

For the test set, the **sensitivity** is $0.83$ and **specificity** is $0.82$
